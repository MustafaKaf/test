## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
df.boxplot(column=['Age'])
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5, color='blue')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')

# Show the plot
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
bubble_size = df['Age']  # You can adjust the multiplication factor for bubble size

# Create a bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_size, alpha=0.5, c='blue')

# Add labels and title
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Bubble Size: Age)')

# Show the plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT DatabaseDesireNextYear, Count(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""

# Assuming you have a connection named "conn"
# conn = ...

# Execute the SQL query and read it into a DataFrame
df1 = pd.read_sql_query(QUERY, conn)
df1.head()

df1.set_index('DatabaseDesireNextYear', inplace=True)

# Define colors and labels for the pie chart
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
labels = df1.index  # Use the index as labels

# Create the pie chart
plt.figure(figsize=(10, 6))  # Set the figure size
plt.pie(df1['Count'], labels=labels, colors=color_list, autopct='%1.1f%%', startangle=90, shadow=True, pctdistance=0.85)
plt.title('Top 5 Database Desire Next Year')  # Add the title
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Add legend with proper labels
plt.legend(labels, loc='upper left')

# Show the pie chart
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

age_filtered = df[(df['Age'] >= 30) & (df['Age'] <= 35)]


median_workweek = age_filtered['WorkWeekHrs'].median()
median_coderev = age_filtered['CodeRevHrs'].median()


plt.figure(figsize=(10, 6))
plt.bar('Age 30-35', median_workweek, label='Median WorkWeekHrs', color='blue')
plt.bar('Age 30-35', median_coderev, label='Median CodeRevHrs', bottom=median_workweek, color='orange')


plt.xlabel('Age Group')
plt.ylabel('Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs (Age 30-35)')
plt.legend()


plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
age_filtered1 = df[(df['Age'] >= 45) & (df['Age'] <= 60)]


median_conv = age_filtered1['ConvertedComp'].median()
median_convertedcomp_by_age = age_filtered1.groupby('Age')['ConvertedComp'].median()


plt.figure(figsize=(12, 6))
plt.plot(median_convertedcomp_by_age.index, median_convertedcomp_by_age.values, marker='o', color='blue', linestyle='-')


plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp by Age (Ages 45-60)')


plt.grid(True)  # Add gridlines
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

mainbranch_counts = df['MainBranch'].value_counts()


plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='blue')


plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Count of Respondents by MainBranch')


plt.show()
```
