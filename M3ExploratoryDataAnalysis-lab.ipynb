The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.

Plot the distribution curve for the column `ConvertedComp`.


```python
# your code goes here
filtered_data = df['ConvertedComp'].dropna()


sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.distplot(filtered_data, bins=30, kde=True, color='blue')


plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Distribution of Converted Compensation')

plt.show()
```

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['ConvertedComp'])

# Add labels and title
plt.xlabel('Converted Compensation')
plt.title('Box Plot of Converted Compensation (Including Outliers)')

# Show the plot
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
filtered_comp = df['ConvertedComp'].dropna()

# Calculate the first quartile (Q1) and third quartile (Q3)
q1 = np.percentile(filtered_comp, 25)
q3 = np.percentile(filtered_comp, 75)

# Calculate the Interquartile Range (IQR)
iqr = q3 - q1

print("Interquartile Range (IQR) for ConvertedComp:", iqr)
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers = (filtered_comp < lower_bound) | (filtered_comp > upper_bound)


num_outliers = outliers.sum()

print("Number of outliers in ConvertedComp column:", num_outliers)
```
